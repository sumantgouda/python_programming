```python
# Write a program to find the euclidean distance between two coordinate
import math
x1=float(input("enter value of x1:"))
x2=float(input("enter value of x2:"))
y1=float(input("enter value of x3:"))
y2=float(input("enter value of x4:"))

distance=math.sqrt((x2-x1)**2 + (y2-y1)**2)

print(f"euclidean distance between two :{distance}")
```
