```python
# Write a program that will reverse a  number.Also it checks whether the reverse is a given number true or false
n=int(input("enter number:"))

      
rev=0
r=n
while(n>0):
    rem=n%10
    rev=rev*10+rem
    n=n//10
print(f"reversed nuber:{rev}")

if r==rev:
    print("True")
else:
    print("False")
```
