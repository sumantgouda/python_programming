```python
# Write a program that will check whether the number is armstrong number or not.
n=int(input("enter number:"))
temp=n
d=len(str(n))
sum=0
while(temp>0):
    digit=temp%10
    sum=sum+digit**d
    temp=temp//10
if sum==n:
    print(f"{n} is armstrong number")
else:
    print(f"{n} is not armstrong number")
```
