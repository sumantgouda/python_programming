```python
# 21. Write a program to find the volume of the cylinder. Also find the cost when ,when the cost of 1 litre milk is 40Rs.
r=float(input("Enter radius of cylinder in cm:"))
h=float(input("Enter height of cylinder in cm:"))
volume=3.142*(r**2)*h
ltr=volume/1000
milk=ltr*40
print(f"Volume of cylinder is: {volume:.2f}")
print(f"cost of {ltr:.2f} litre milk is: ${milk:.2f}")
```
