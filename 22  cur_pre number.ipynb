```python
# 22. Print the sum of the current number and the previous number Write a program to iterate the first 10 numbers and in each iteration, print the 
# sum of the current and previous number
n=int(input("Enter number:"))
p_num=0
for i in range(n+1):
    x=p_num+i
    p_num=i
    print(x,end=" ")
```
